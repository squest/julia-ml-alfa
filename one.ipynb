```julia
function sumsieve2(lim::Int)
    res::Int,primes::Array{Bool},llim::Int=2,trues(lim),isqrt(lim)
    bates::Int = iseven(llim) ? llim+1:llim+2
    for i = 3:2:llim
        if primes[i]
            for j = i*i:i+i:lim
                primes[j] = false
            end
            res += i
        end
    end
    for i = bates:2:lim
        if primes[i]
            res += i
        end
    end
    res
end
```

```julia
function sumsieve(lim::Int)
    res :: Int, primes::Array{Bool}, llim :: Int = 2,trues(lim),isqrt(lim)
    for i in 3:2:lim
        if i <= llim
            if primes[i]
                for j in i*i:i+i:lim
                    primes[j] = false
                end
                res += i
            end
        else
            if primes[i]
                res += i
            end
        end
    end
    res
end
```

```julia
function sol10(lim::Int)
    primes :: Array{Bool} = trues(lim)
    llim :: Int = isqrt(lim)
    res :: Int = 2
    for i in 3:2:lim
        if i <= llim 
            if primes[i]
                for j in i*i:i*2:lim
                    primes[j] = false
                end
                res += i
            end
        else
            if primes[i]
                res += i
            end
        end
    end
    return res
end
```
